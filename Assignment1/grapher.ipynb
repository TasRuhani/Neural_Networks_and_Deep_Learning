```python
import matplotlib.pyplot as plt
from graph_data import graph_data


def plot_metric(hidden_layers, metric_key, title):
    plt.figure()

    for key in graph_data:
        if key.endswith(f"_{hidden_layers}"):
            epochs = range(1, len(graph_data[key][metric_key]) + 1)
            plt.plot(epochs, graph_data[key][metric_key], label=key)

    plt.xlabel("Epoch")
    plt.ylabel(metric_key.replace("_", " ").title())
    plt.title(f"{title} (Hidden Layers = {hidden_layers})")
    plt.legend()
    plt.grid(True)
    plt.show()


def generate_all_graphs():
    # Extract hidden layer numbers and sort them numerically
    hidden_layer_values = sorted(
        {int(key.split("_")[-1]) for key in graph_data.keys()}
    )

    # Now it will always go 5 -> 10 -> 20
    for hl in hidden_layer_values:
        plot_metric(hl, "train_loss", "Training Loss vs Epoch")
        plot_metric(hl, "val_loss", "Validation Loss vs Epoch")
        plot_metric(hl, "train_acc", "Training Accuracy vs Epoch")
        plot_metric(hl, "val_acc", "Validation Accuracy vs Epoch")


if __name__ == "__main__":
    generate_all_graphs()
```
